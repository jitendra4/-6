```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv('D:\\GooglePlayStoreApps.csv', nrows=1000)
df.head()
```

```python
df.shape
```

```python
x=df['Rating']
y=df["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title('google play store')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c= 'y', marker='*',s= 100, alpha= 0.5, linewidth=5, edgecolor='g', )#verts= '<')
plt.title('google play store')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c= 'y', marker='*',s= 100, alpha= 0.5, linewidth=10, edgecolor='g', )#verts= '<')

plt.scatter(x,df["Installs"], c= 'm', marker='<',s= 100, alpha= 0.5, linewidth=10, edgecolor='c', )#verts= '<')
plt.title('google play store')
plt.xlabel('Rating')
plt.ylabel('Reviews and install')
plt.show()
```
